### Modelimizi oluşturmaya başlayalım

Önce veriyi test ve train için ayıralım

```python
y=dataFrame["Type"].values
x=dataFrame.drop("Type",axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.3,random_state=15)
```

Küçültme işlemlerimizi yapalım(scale)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

fit işlemlerini yapalım

```python
scaler.fit(x_train)
```

```python
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

EarlyStopping ve Dropout'u overfitting problemlerini çözmek için kullanacağız

```python
x_train.shape
```

```python
model=Sequential()

model.add(Dense(units=30,activation="relu"))
#genelde şöyle bir yaklaşım vardır(duruma göre değişebilir) kaç tane column'umuz varsa input layerınızı ona eşit yapmanız önerilir
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=15,activation="relu"))
#30 ile 1 arasına da arada bir değer koyabiliriz
model.add(Dense(units=1,activation="sigmoid"))
#sigmoid 0 ve 1 arasında bir değer veriyordu.
#sınıflandırma problemlerinde çıkış layerına sigmoid verebiliriz

model.compile(loss="binary_crossentropy", optimizer="adam")
```

train edelim
* epocs: bizim veri setimizin üzerinden kaç kere geçip train edeceğini gösteren rakamdır. Çok abartmak overfittinge yol açabilir.

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test,y_test),verbose=1)
```

```python
modelKaybi=pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

Böyle bir durumla karşılaşırsak şunu diyebiliriz eğer val_los saçma sapan artmaya başlarsa epoch'u durdur diyebiliriz. Bunun için de earlyStopping kullanacağız

```python
model=Sequential()

model.add(Dense(units=30,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=1,activation="sigmoid"))

model.compile(loss="binary_crossentropy", optimizer="adam")
```

```python
earlyStopping=EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test,y_test),verbose=1,callbacks=[earlyStopping])
```

```python
modelKaybi=pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

Dropout layer özelliğini kullanalım. Çok fazla layerlarla ilgili overfitting yaşıyorsak training süresince modelimizden random bir şekilde atılacak.

```python
model=Sequential()

model.add(Dense(units=30,activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15,activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15,activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(units=1,activation="sigmoid"))

model.compile(loss="binary_crossentropy", optimizer="adam")
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test,y_test),verbose=1,callbacks=[earlyStopping])
```

```python
kayipDf=pd.DataFrame(model.history.history)
```

```python
kayipDf.plot()
```

Overfittingin biraz üstesinden geldik. Dropout kullanınca lossların zigzag şeklinde olması normal.

```python
tahminlerimiz = model.predict_classes(x_test)
```

```python
tahminlerimiz
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,tahminlerimiz))
```

precision: ne kadar duyarlı bir şekilde tahmin ettiğini belirtir.
0'ları %89 1'leri %91 oranında düzgün bir şekilde tahmin ettiğini söylüyor.

```python
print(confusion_matrix(y_test,tahminlerimiz))
```
